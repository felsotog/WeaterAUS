# CRISP-DM 6: Despliegue

Algunas de las variables en este conjunto de datos son categóricas y representan diferentes características climáticas, como la dirección del viento, la ubicación de la estación meteorológica y la cantidad de lluvia.

Estas variables categóricas podrían ser transformadas en variables dummy para permitir su uso en modelos estadísticos o de aprendizaje automático. Por ejemplo, la variable "WindDir9am" representa la dirección del viento a las 9 de la mañana y puede tener valores como "N", "NE", "E", "SE", "S", "SW", "W" o "NW".

Para convertir esta variable en una variable dummy, se crearían ocho nuevas variables binarias, una para cada dirección del viento, con valores 0 o 1 que indican si esa dirección del viento está presente o no en esa fila de datos. De esta manera, la información contenida en la variable original se puede utilizar para analizar y predecir el clima, sin tener que asignar números arbitrarios a las diferentes direcciones del viento.

Además, esta técnica también podría aplicarse a otras variables categóricas en el conjunto de datos, como la variable "Location" que representa la ubicación de la estación meteorológica, o la variable "RainToday" que indica si ha llovido o no en el día actual.

En resumen, transformar variables string en variables dummy en el conjunto de datos "weatherAus" permite su uso en modelos estadísticos o de aprendizaje automático. Esta técnica puede aplicarse a varias variables categóricas en el conjunto de datos, mejorando así la calidad del análisis y la precisión de las predicciones.

Habiendo hecho esta división probamos  3 modelos diferentes:
Arbol de Decisión
K-Vecinos mas cercanos y
Bayes Ingenuos

En el caso del algoritmo de los K-Vecinos más cercanos, este requiere que utilizemos gran parte del dataset para entregar resultados consistentes, pero aun así son resultados de baja precisión.
Para el Arból de Decisión los resultados de precisión 79.81%.
Y para el algoritmo de Bayes Ingenuos la precisión 65.9%

Por lo que, para nuestro dataset, el algoritmo de predicción que recomendamos utilizar es el algoritmo de Bayes Ingenuos.
